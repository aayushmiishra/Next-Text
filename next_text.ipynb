```python
with open('blue_castle.txt', 'r') as file:
    kitaab = file.read()
```

```python
print(kitaab)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([kitaab])
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []

for sentence in kitaab.split('\n'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1, len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len=max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequnces=pad_sequences(input_sequences, maxlen=max_len, padding='pre')
```

```python
padded_input_sequnces
```

```python
X=padded_input_sequnces[:, :-1]
```

```python
y=padded_input_sequnces[:, -1]
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y, num_classes=8727)
```

```python
y.shape
```

```python
# for Architecture now:
# Step 1 = EMBEDDING
# Step 2 = LSTM
# Step 3 = DENSE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model=Sequential()
model.add(Embedding(8727, 100))
model.add(LSTM(150))
model.add(Dense(8727, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.build(input_shape=(None, max_len))
model.summary()
```

```python
model.fit(X,y,epochs=100  )
```
